```python
""" Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example. """
Ans:
    In pandas, a DataFrame is a two-dimensional data structure that resembles a table or a spreadsheet. It consists of rows and columns, where each column can contain different data types (e.g., integers, floats, strings). 
    On the other hand, a pandas Series is a one-dimensional labeled array capable of holding any data type. It can be seen as a single column of a DataFrame or a standalone data structure.
```

```python
#Example :
d = [1,2,3,4,5]
pd.Series(d)
```

```python
""" Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions? """

Ans:
    Pandas provides a wide range of functions to manipulate and transform data in a DataFrame.
Here are some common functions:
1.head() and tail(): These functions allow you to view the first or last few rows of a DataFrame, respectively. 
2.describe(): This function provides descriptive statistics of the DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartile values
3.shape: This attribute returns the dimensions (number of rows and columns) of the DataFrame.
4.sort_values(): This function allows you to sort the DataFrame by one or more columns. You can specify ascending or descending order for each column.
5.groupby(): This function enables you to group the data based on one or more columns and perform aggregate operations on the grouped data.

These are just a few examples of the many functions available in pandas for data manipulation. The choice of function depends on the specific requirements and operations you want to perform on the DataFrame.
```

```python
""" Q6. Which of the following is mutable in nature Series, DataFrame, Panel? """
Ans:
    In pandas, both Series and DataFrame are mutable in nature, meaning you can modify their contents after creation. On the other hand, Panel objects are deprecated and no longer recommended for use as of pandas version 0.25.0. Instead, the preferred way to handle three-dimensional data is by using the MultiIndex DataFrame. Therefore, Series and DataFrame are the main mutable data structures in pandas.
```
