```python
"""Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib."""
ans:
    Matplotlib is a widely-used data visualization library in Python. It provides a comprehensive range of functions for creating various types of static, interactive, and animated visualizations.
use:
     Matplotlib is commonly used for creating graphs, charts, and plots to visually represent data in a clear and understandable manner.
The primary reasons for using matplot library are:
    1. Data visualization
    2.Flexibility
    3.Publication quality plots
    4.compatibility
    5.Extensibility
The five plots are:
    1.Line Plot: A basic plot used to display data points connected by straight lines. It is often used to show trends over time or relationships between variables.
    2.Bar Plot: A plot where data is represented using rectangular bars. It is useful for comparing categorical data or showing the frequency distribution of different categories.
    3.Histogram: A graphical representation of the distribution of numerical data. It consists of bins that represent ranges of data values and shows how data is distributed within those ranges.
    4.Scatter Plot: A plot that displays individual data points as dots in a two-dimensional space. It is useful for visualizing the relationship between two continuous variables.
    5.Pie Chart: A circular chart divided into slices that represent proportions of a whole. It is often used to show percentages or parts of a whole.
```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot."""
Ans:
    Scatter Plot: A plot that displays individual data points as dots in a two-dimensional space. It is useful for visualizing the relationship between two continuous variables.
```

```python
#code:
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""

Ans:
    The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to divide the figure into a grid of rows and columns and specify which grid cell a new plot should occupy.
The primary reasons for using the subplot() function are:
    1.comparative analysis
    2.Contextualization
    3.Storytelling
    4.optimizing space
```

```python
#For line1:
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("For line 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
""" Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])"""
Ans:
    Bar plot:
        A bar plot, also known as a bar chart, is a graphical representation of data using rectangular bars of varying lengths. Each bar corresponds to a specific category or group, and the length of the bar is proportional to the value or frequency associated with that category.
used for:
    Bar plots are commonly used to visualize and compare data across different categories or groups.
```

```python
#Vertical bar plot:
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("companies")
plt.ylabel("profits")
plt.title("Bar plot")
plt.show()
```

```python
#horizontal bar plot:
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("companies")
plt.ylabel("profits")
plt.title("Bar plot")
plt.show()
```

```python
""" Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200) """
Ans:
    Box plot:
        A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a compact summary of the distribution of a dataset. It displays key measures of the dataset's central tendency, spread, and potential outliers.
    used for:
        A box plot is particularly useful for identifying the range, variability, and skewness of the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.title("Box Plot-1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.title("Box Plot-2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
